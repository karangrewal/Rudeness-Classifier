```python
data = []
```

```python
result = subprocess.run(['java -cp "/home/dluo/Downloads/stanford-corenlp-full-2017-06-09/*" -mx5g  edu.stanford.nlp.sentiment.SentimentPipeline -file ' + datadir + 'rude-12.txt'], shell=True, stdout=subprocess.PIPE)
result.stdout
```

```python
for file in os.listdir(datadir):
    process = 'java -cp "/home/dluo/Downloads/stanford-corenlp-full-2017-06-09/*" -mx5g edu.stanford.nlp.sentiment.SentimentPipeline -file ' + datadir + file
    result = subprocess.run([process], shell=True, stdout=subprocess.PIPE)
    data.append([file, result.stdout])
    print(file, 'done')
```

```python
data
```

```python
data_split = [[file, entry.decode("utf-8").split('\n')] for file, entry in data]
```

```python
data_split
```

```python
df = pd.DataFrame(columns=('File', 'Sentence', 'Sentiment'))
for file, entry in data_split:
    for i in range(0, len(entry)-1, 2): #omit the last one which is just ''
        #print(entry[i], entry[i+1][2:])
        df.loc[len(df)] = [file, entry[i], entry[i+1][2:]] #[2:] is to get rid of the 2 annoying spaces at the front get it together stanford
```

```python
df
```

```python
df.to_csv('Stanford_NLP_Sentence_Sentiment_Analysis.csv')
```
