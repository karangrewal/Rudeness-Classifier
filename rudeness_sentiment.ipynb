```python
data = []
```

```python
for file in os.listdir(datadir):
    process = 'java -cp "*" -mx5g edu.stanford.nlp.sentiment.SentimentPipeline -file ' + datadir + file
    result = subprocess.run([process], shell=True, stdout=subprocess.PIPE)
    data.append(result.stdout)
    print(file, 'done')
```

```python
data
```

```python
data_split = [entry.decode("utf-8").split('\n') for entry in data]
```

```python
data_split
```

```python
df = pd.DataFrame(columns=('Sentence', 'Sentiment'))
for entry in data_split:
    for i in range(0, len(entry)-1, 2): #omit the last one which is just ''
        #print(entry[i], entry[i+1][2:])
        df.loc[len(df)] = [entry[i], entry[i+1][2:]] #[2:] is to get rid of the 2 annoying spaces at the front get it together stanford
```
